```python
fig, ax = plt.subplots(1,3, figsize= (12,4))

ax[0].plot([1,3,6],[1,2,4],"r")
ax[0].set_title("First")

ax[1].plot([2,4,6],[1,3,5],"g")
ax[1].set_title("Second")

ax[2].plot([1,2,3],[1,2,3],"y")
ax[2].set_title("Third")

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(2,2, figsize = (8,6))

ax[0,0].plot([1,4,6,8],[1,2,5,7],"b")
ax[0,0].set_title("First One")

ax[0,1].plot([1,2,3,4],[1,2,3,4],"r")
ax[0,1].set_title("Second One")

ax[1,0].plot([1,3,6,8],[1,2,4,7],"g")
ax[1,0].set_title("Third One")

ax[1,1].plot([1,2,3,4],[4,4,4,4],"y")
ax[1,1].set_title("Fourth One")

plt.tight_layout()

plt.show()
```

```python
x = ["Bilal","Tharun","Nibal","Vishnu"]

x1 = np.array([50,45,46,55])
x2 = np.array([60,30,35,40])
x3 = np.array([50,25,30,35])
x4 = np.array([55,40,45,50])

plt.bar(x,x1,color="y")
plt.bar(x,x2,bottom=x1,color="g")
plt.bar(x,x3,bottom=x1+x2,color="r")
plt.bar(x,x4,bottom=x1+x2+x3,color="b")


plt.xlabel("Students")
plt.ylabel("Score")
plt.legend(["First","Second","Third","Fourth"])

plt.title("Marks of Students in 4 exams")

plt.show()
```
