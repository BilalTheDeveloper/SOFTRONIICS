```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set(style="dark")
d = np.random.normal(200,20,100)

sns.histplot(d,kde=True,color="r")
```

```python
sns.set(style="dark")
iris = sns.load_dataset("iris")

ax = sns.stripplot(x='species',y='sepal_length',data=iris)
plt.title("FIRST")
plt.show()
```

```python
sns.set(style="white")
penguins = sns.load_dataset("penguins")
sns.kdeplot(data=penguins, x="flipper_length_mm", hue="species", multiple="stack")
```

```python
x = ["jan","feb","mar","apr","may","jun","jul","aug"]
y = [20000,30000,20222,50000,10000,6000,24000,32102]

ax = sns.stripplot(x=x,y=y)
ax.set(xlabel="Months",ylabel="Spend Amount")
plt.title("Report")
plt.show()
```
