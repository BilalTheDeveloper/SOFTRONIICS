```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.array([1,2,3,4,5])

y = np.array([2,4,6,8,10])

plt.xlabel("X-Axis")

plt.ylabel("Y-Axis")

plt.title("Basic Line Plot")

plt.plot(x,y)

plt.show()
```

```python
x = np.array([0,1,2,3,4])

y = np.array([0,1,4,9,16])

plt.xlabel("X-Axis")

plt.ylabel("Y-Axis")

plt.title("Basic Line Plot")

plt.plot(x,y,marker = "o")

plt.show()
```

```python
x = np.array([1,2,3,4])
y = np.array([1,4,9,16])

plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.plot(x,y,linestyle = "dashed",label = "First")

x = ([1,2,3,4])
y1 = ([2,3,6,8])

plt.plot(x,y1,linestyle = "solid",label = "Second")

plt.legend()

plt.show()
```

```python
x = np.array(["A","B","C","D"])
y = np.array([5,7,3,8])

plt.xlabel("Categories")
plt.ylabel("Values")

plt.bar(x,y)
plt.show()
```

```python
x = np.array(["E","F","G","H"])
y = np.array([8,4,6,3])

plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Horizontal Bar Plot")

plt.barh(x,y)
plt.show()
```

```python
x = np.array(["X","Y","Z"])

y = np.array([7,2,5])

plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Annot")

plt.bar(x,y,)


plt.show()
```
