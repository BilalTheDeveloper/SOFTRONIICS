```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

df = pd.read_csv("A.csv")
df['Date'] = pd.to_datetime(df['Date'])  
df = df.sort_values('Date')  


plt.figure(figsize=(10, 5))
plt.plot(df['Date'], df['Close'], label='Stock Price', color='blue')
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.title("Stock Prices Over Time")
plt.legend()
plt.grid(True)
plt.show()


months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
city1 = [2, 3, 7, 13, 18, 22, 25, 24, 20, 14, 8, 3]  
city2 = [-5, 0, 5, 12, 17, 21, 23, 22, 17, 10, 2, -3]  
city3 = [10, 12, 16, 20, 25, 30, 33, 32, 28, 22, 16, 12] 

plt.figure(figsize=(10, 5))
plt.plot(months, city1, marker='o', label="City A", linestyle='dashed')
plt.plot(months, city2, marker='s', label="City B", linestyle='dotted')
plt.plot(months, city3, marker='^', label="City C", linestyle='solid')
plt.xlabel("Month")
plt.ylabel("Temperature (°C)")
plt.title("Monthly Average Temperatures of Three Cities")
plt.legend()
plt.grid(True)
plt.show()


sns.set_style("darkgrid")
plt.figure(figsize=(10, 5))
plt.plot(df['Date'], df['Close'], label="Stock Price", color="blue")
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.title("Stock Prices with Darkgrid Style")
plt.legend()
plt.show()

sns.set_style("whitegrid")
plt.figure(figsize=(10, 5))
plt.plot(df['Date'], df['Close'], label="Stock Price", color="blue")
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.title("Stock Prices with Whitegrid Style")
plt.legend()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data1 = {
    'height': [150, 160, 170, 180, 190, 155, 165, 175, 185, 195],
    'weight': [50, 60, 65, 80, 90, 52, 63, 78, 85, 95],
    'gender': ['Male', 'Male', 'Male', 'Male', 'Male', 'Female', 'Female', 'Female', 'Female', 'Female']
}
df1 = pd.DataFrame(data1)


sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df1, x='height', y='weight', hue='gender', style='gender', s=100)
plt.title('Height vs Weight by Gender')
plt.show()
```

```python
data2 = {
    'square_footage': [800, 900, 1000, 1100, 1200, 1300, 1400],
    'price': [150000, 160000, 180000, 200000, 220000, 240000, 260000],
    'location': ['Urban', 'Urban', 'Suburban', 'Suburban', 'Rural', 'Rural', 'Urban'],
    'bedrooms': [2, 3, 3, 4, 2, 4, 5]
}
df2 = pd.DataFrame(data2)

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df2, x='square_footage', y='price', hue='location', size='bedrooms', sizes=(50, 300))
plt.title('House Prices by Square Footage, Location, and Bedrooms')
plt.show()
```

```python
sns.set_context("talk")

plt.figure(figsize=(10, 7))
sns.scatterplot(data=df2, x='square_footage', y='price', hue='location', size='bedrooms', sizes=(100, 500))
plt.title('House Prices by Square Footage, Location, and Bedrooms')
plt.xlabel("Square Footage")
plt.ylabel("Price")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python


iris = sns.load_dataset('iris')

plt.figure(figsize=(8, 6))
sns.regplot(data=iris, x='sepal_length', y='petal_length', ci=None)
plt.title('Sepal Length vs Petal Length (No Confidence Interval)')
plt.show()
```

```python

plt.figure(figsize=(16, 5))

plt.subplot(1, 2, 1)
sns.regplot(data=iris, x='sepal_length', y='petal_length', ci=68)
plt.title('Confidence Interval: 68%')

plt.subplot(1, 2, 2)
sns.regplot(data=iris, x='sepal_length', y='petal_length', ci=99)
plt.title('Confidence Interval: 99%')

plt.tight_layout()
plt.show()
```

```python
tips = sns.load_dataset('tips')

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

sns.regplot(data=tips, x='total_bill', y='tip', ax=axes[0])
axes[0].set_title('Total Bill vs Tip')

sns.regplot(data=tips, x='size', y='tip', ax=axes[1])
axes[1].set_title('Party Size vs Tip')

plt.tight_layout()
plt.show()
```

```python


mpg = sns.load_dataset('mpg')

mpg = mpg.dropna(subset=['displacement', 'horsepower'])

palettes = ['Blues', 'Reds', 'coolwarm']

plt.figure(figsize=(18, 5))
for i, palette in enumerate(palettes):
    plt.subplot(1, 3, i + 1)
    sns.scatterplot(data=mpg, x='horsepower', y='displacement', palette=palette, hue='origin')
    plt.title(f'Palette: {palette}')

plt.tight_layout()
plt.show()
```

```python
# Task 4.2: Style: white with despine
sns.set_style('white')
plt.figure(figsize=(6, 5))
sns.scatterplot(data=mpg, x='horsepower', y='displacement', hue='origin')
sns.despine()
plt.title('White Style with Despine')
plt.show()

# Style: ticks with despine
sns.set_style('ticks')
plt.figure(figsize=(6, 5))
sns.scatterplot(data=mpg, x='horsepower', y='displacement', hue='origin')
sns.despine()
plt.title('Ticks Style with Despine')
plt.show()

```

```python
# Load the iris dataset
iris = sns.load_dataset('iris')

# Custom color palette
custom_palette = sns.color_palette(['#1f77b4', '#ff7f0e', '#2ca02c'])

plt.figure(figsize=(7, 6))
sns.scatterplot(data=iris, x='sepal_length', y='petal_length', hue='species', palette=custom_palette)
plt.title('Custom Palette for Iris Species')
plt.show()
```

```python


tips = sns.load_dataset('tips')

g = sns.FacetGrid(tips, col="time", row="day", margin_titles=True)
g.map(sns.scatterplot, "total_bill", "tip")
g.fig.subplots_adjust(top=0.9)
g.fig.suptitle('Total Bill vs Tip by Day and Time')
plt.show()
```

```python
# Load the iris dataset
iris = sns.load_dataset('iris')

# Task 5.2: FacetGrid scatterplot for sepal_width vs petal_width with hue by species
g = sns.FacetGrid(iris, hue="species", height=4, aspect=1.5)
g.map(sns.scatterplot, "sepal_width", "petal_width").add_legend()
g.fig.suptitle('Sepal Width vs Petal Width by Species', y=1.03)
plt.show()
```

```python
# Task 5.3: FacetGrid with regression lines for each species
g = sns.FacetGrid(iris, col="species", height=4, aspect=1.2)
g.map(sns.regplot, "sepal_length", "petal_length", scatter_kws={"s": 30}, ci=None)
g.fig.subplots_adjust(top=0.85)
g.fig.suptitle('Sepal Length vs Petal Length by Species (Regression Lines)')
plt.show()
```
